```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
np.random.seed(20)
var1 = np.random.randint(0, 10, 10) # dias de vacaciones
var2 = var1 + np.random.normal(0, 1, 10) # % dinero gastado
```

```python
sns.regplot(x = var1, y = var2, ci = 80)
plt.xlabel('Dias de vacaciones')
plt.ylabel('% dinero gastado')
plt.show()
```

```python
tup = list(zip(var1, var2))
df = pd.DataFrame(tup, columns = ["dias", "dinero"])
```

### Calculo de intervalos

Si n<30 usamos st.t.interval

Si n>=30 usamos st.norm.interval


```python
ci_x, ci_y = stats.t.interval(df=len(df) - 1, loc=np.mean(df, axis=0), scale=stats.sem(df), confidence=0.95)

print("Intervalo de confianza para x: ", round(ci_x[0], 2), " - ", round(ci_x[1], 2))
print("Intervalo de confianza para y: ", round(ci_y[0], 2), " - ", round(ci_y[1], 2))
```

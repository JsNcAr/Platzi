```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[-7, 1], [-3, 1]])
b = np.array([[2], [5]])

x = np.linalg.solve(A, b)
print(x)
```

```python
x = np.linspace(-10, 10, 100)

y1 = 7 * x + 2
y2 = 3 * x + 5

plt.plot(x, y1)
plt.plot(x, y2)
plt.grid(True)
plt.xlim(-2, 2)
plt.ylim(0, 8)
plt.show()
```

```python
v1 = np.array([1/2,1/2,1/2,1/2])
# norma 2
print(np.linalg.norm(v1))
```
